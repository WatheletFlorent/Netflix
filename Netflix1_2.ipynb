```python
#11. Étudier les catégories avec le plus de contenu

donnes["listed_in"].head(10)

", ".join(donnes["listed_in"].dropna())

categories = ", ".join(donnes["listed_in"].dropna()).split(", ")
categories =pd.Series(categories)
print(categories.value_counts().head())
```

```python
#12. Afficher les directeurs qui ont produit le plus de films/sé-ries disponibles sur Netflix

donnes["director"].head(10)

", ".join(donnes["director"].dropna())

categories = ", ".join(donnes["director"].dropna()).split(", ")
categories =pd.Series(categories)
print(categories.value_counts().head())
```

```python
#13. Voir si Jan Suter travaille souvent avec les mêmes acteurs

loc = donnes.loc[donnes.director.notna()]
donnes_jan_suter = loc.loc[loc.director.str.contains('Jan Suter')]

categories = ", ".join(donnes_jan_suter["cast"].dropna()).split(", ")
categories =pd.Series(categories)
print(categories.value_counts().head())
```

```python
#14. Représenter les dix pays qui ont produit le plus de contenus disponibles sur Netflix,
#avec le nombre de contenus par pays

import matplotlib.pyplot as plt
import seaborn as sns

pays = donnes['country'].value_counts().head()
defpays = donnes.loc[donnes.country.isin(pays.index)]
sns.countplot(y='country', data=defpays)
plt.xticks(rotation=90)
```

```python
#15. Tracer un graphe à barres du nombre de films/séries parclassement de contenu (rating)

import matplotlib.pyplot as plt
import seaborn as sb

plt.figure(figsize=(10,5))
sb.countplot(x='rating', data= donnes)
plt.show()
```

```python
#16. Afficher l’évolution du nombre de films/séries disponiblessur Netflix au cours du temps

data=donnes[donnes['date_added'].notna()]
date_year= pd.to_datetime(data['date_added']).dt.year

donnes['year_added']=date_year
data_group=donnes.groupby(['type','year_added']).size().reset_index(name='Count')

plt.figure(figsize=[15,5])
plt.xticks(rotation=90)
sns.pointplot('year_added', 'Count', hue = 'type', data=data_group)
```

```python
#17. Afficher la distribution de la durée des films disponiblessur Netflix

movie = donnes[(donnes["type"]=="Movie")]
duree = pd.Series(movie['duration']).str.replace("min", "").astype("int").sort_values(ascending=False)
sb.distplot(duree)
```

```python
# 18. Tracer un graphique représentant le nombre de séries par modalité de nombre de saisons

movie = donnes[(donnes["type"]=="TV Show")]
plt.figure(figsize=(10, 6))
sns.countplot(x='duration', data = movie)
plt.xticks(rotation=90)
```

```python
data = pd.read_csv("/home/wathelet/projet_data/Pandas_Netflix/netflix_titles.csv")

tmp2 = data.loc[data["listed_in"].notna()][['show_id', 'listed_in']]

show_listed_in = pd.DataFrame(columns = ['show_id', 'listed_in'])


for i in range(tmp2.shape[0]):
    tmp = tmp2.iloc[i]["listed_in"].split(", ")
    show_id = [tmp2.iloc[i]["show_id"]]*len(tmp)
    df = pd.DataFrame({'show_id': show_id,
                   'listed_in': tmp}, columns = ['show_id', 'listed_in'])
    show_listed_in = show_listed_in.append(df, ignore_index=True)

show_listed_in
```
